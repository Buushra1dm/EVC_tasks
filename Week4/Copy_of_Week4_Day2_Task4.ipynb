```python
import pandas as pd
import torch
from torch.utils.data import Dataset
from transformers import AutoTokenizer

class CustomMaskedDataset(Dataset):
    def __init__(self, file_path, tokenizer_name='bert-base-uncased', max_length=128):
        # Load the data
        self.data = pd.read_csv(file_path)
        self.tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        # Get the text
        text = self.data.iloc[idx]['text']

        # Tokenize the text
        encoding = self.tokenizer(
            text,
            padding='max_length',
            truncation=True,
            max_length=self.max_length,
            return_tensors='pt'
        )

        # Return the tokenized text as a dictionary
        return {
            'input_ids': encoding['input_ids'].squeeze(),  # remove batch dimension
            'attention_mask': encoding['attention_mask'].squeeze(),  # remove batch dimension
            'labels': encoding['input_ids'].squeeze()  # For MLM, labels are the same as input_ids
        }
```

```python
import pandas as pd

# Define the data
data = {
    'text': [
        "The quick brown [MASK] jumps over the lazy dog.",
        "A [MASK] day in the park is wonderful.",
        "She sells sea [MASK] by the seashore."
    ]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Save DataFrame to CSV
csv_file_path = 'masked_sentences.csv'
df.to_csv(csv_file_path, index=False)

print(f"CSV file created at: {csv_file_path}")
```

```python
from torch.utils.data import DataLoader

# Initialize the dataset
dataset = CustomMaskedDataset(file_path='/content/masked_sentences.csv')

# Initialize the DataLoader
dataloader = DataLoader(dataset, batch_size=8, shuffle=True)

# Example of how to iterate through the dataloader
for batch in dataloader:
    input_ids = batch['input_ids']
    attention_mask = batch['attention_mask']
    labels = batch['labels']
    print(input_ids)
    print(attention_mask)
    print(labels)
```
